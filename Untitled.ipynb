```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
cancer = load_breast_cancer()
```

```python
features = list(cancer.feature_names)
features.append('target')
```

```python
print(cancer.target_names)
```

```python
features
```

```python
cancer_data = pd.DataFrame(np.c_[cancer.data, cancer.target], columns = features)
```

```python
cancer_data
```

```python
cancer_data.shape
```

```python
#line plot for avereage radius and the average area of the tumor
plt.figure(figsize = (16,6))
sns.lineplot(data = cancer_data['mean radius'], label = "Radius")
sns.lineplot(data = cancer_data['mean area'], label = "Area")
plt.title("Tumor detail for classification of type")
```

```python
#Barchart for tumor area and target
plt.figure(figsize = (14,5))
sns.barplot(x = cancer_data['target'], y = cancer_data['worst symmetry'])
plt.title("Barchart Area of tumor vs target")
```

```python
#Scatter plot for area vs target
plt.figure(figsize = (14,6))
sns.scatterplot(x = cancer_data['mean area'], y = cancer_data['worst compactness'], hue = cancer_data.target)
```

```python
plt.figure(figsize = (14,5))
sns.scatterplot(x = cancer_data['mean radius'], y = cancer_data['worst smoothness'], hue = cancer_data.target)
```

```python
plt.figure(figsize=(16,7))
sns.heatmap(data = cancer_data.corr(), annot = True)
```

```python
sns.pairplot(cancer_data, hue = "target", vars = ['mean radius', 'mean texture', 'mean concavity','mean perimeter', 'mean area','mean compactness','mean smoothness', 'mean concave points','mean symmetry'])
```

### Splitting data 

```python
X = cancer_data.drop('target', axis = 1)
y = cancer_data['target']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X,y, test_size = 0.4, random_state = 0)
```

```python
X_train
```

### Defining  Model

```python
#kernel Polynomial
model = SVC(kernel = 'poly', degree = 2, gamma = 'auto')
```

```python
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_val)
```

```python
score = accuracy_score(y_val, preds)
print("The accuracy of the model for this dataset is:%f"%score)
```
